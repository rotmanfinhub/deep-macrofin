```python
from deepMacroFin.models import Agent
import matplotlib.pyplot as plt
```

```python
config = {
    "device": "cpu",
    "hardcode_function": lambda x: x ** 2,
    "derivative_order": 2
}
model = Agent("qa", ["x"], config)
fig, ax = plt.subplots(1, 3, figsize=(20, 6))
model.plot("qa", {"x": [-1, 1]}, ax=ax[0])
model.plot("qa_x", {"x": [-1, 1]}, ax=ax[1])
model.plot("qa_xx", {"x": [-1, 1]}, ax=ax[2])
plt.tight_layout()
plt.show()
```

```python
config = {
    "device": "cpu",
    "hardcode_function": lambda x: x[:, 0] * x[:, 1],
    "derivative_order": 2
}
model = Agent("qa", ["x", "y"], config)
fig, ax = plt.subplots(1, 3, figsize=(20, 6), subplot_kw={"projection": "3d"})
model.plot("qa", ax=ax[0])
model.plot("qa_x", ax=ax[1])
model.plot("qa_y", ax=ax[2])
plt.tight_layout()
plt.show()
```
